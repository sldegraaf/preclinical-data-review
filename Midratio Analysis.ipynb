```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

filename = "datasets/Bishayee Colony Counts 10.27.97-3.8.01.xlsx"
df = pd.read_excel(filename, index_col = 1, skiprows = [0,1])
df.as_matrix()
# deal with the missing value
df = df[df['col3'] != ' ']
# get only the three columns
df_triples = df[['col1', 'col2', 'col3']].apply(pd.to_numeric)
```

```python
# function to find midratio for each row
def midratio(r):
    s = sorted(r)
    low = float(s[0])
    mid = float(s[1])
    high = float(s[2])
    if (high-low) == 0: 
        mr = 'NaN'
    else:
        mr = (mid-low)/(high-low)
    return mr

# compute midratios for all rows in dataset
df_midratio = df_triples.apply(midratio, axis = 1)
df_mr = df_midratio[df_midratio != 'NaN']

# plot a histogram of midratios
plt.figure(figsize=(8, 6))
plt.hist(df_mr, bins = 15, color='r', alpha=0.5)
plt.title('Midratio Values for RTS Colony Counts')
plt.xlabel('Midratio Value')
plt.ylabel('Count')
```

```python
# repeat midratio analysis for Other Investigator Data
filename = "datasets/Other Investigators in Lab.Colony Counts.4.23.92-11.27.02.xlsx"
df2 = pd.read_excel(filename, index_col = 1, skiprows = [0])
df2.as_matrix()
df2 = df2[df2['col3'] != ' ']
df2[['col1', 'col2', 'col3']].head(10)
df2_triples = df2[['col1', 'col2', 'col3']].apply(pd.to_numeric)
df2_midratio = df2_triples.apply(midratio, axis = 1)
df2_mr = df2_midratio[df2_midratio != 'NaN']

plt.figure(figsize=(8, 6))
plt.hist(df2_mr, bins = 15, color='r', alpha=0.5)
plt.title('Midratio Values for Other Investigators Colony Counts')
plt.xlabel('Midratio Value')
plt.ylabel('Count')
```
